```python
customer_df_location=customer_df[['CustomerEasting','CustomerNorthing','CustomerPopulation']]
customer_df_location
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=100, random_state=42, n_init=10)
customer_df_location['Cluster'] = kmeans.fit_predict(customer_df_location)
```

```python
customer_df_location.groupby('Cluster').count()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(customer_df_location["CustomerEasting"], customer_df_location["CustomerNorthing"], c=customer_df_location["Cluster"], cmap="viridis", s=50, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.xlabel("Customer Easting")
plt.ylabel("Customer Northing")
plt.title("K-Means Clustering (k=100)")
plt.legend()
plt.show()
```

```python
from k_means_constrained import KMeansConstrained
```

```python
# Apply KMeansConstrained with k=100
num_clusters = min(100,120)  # Ensure clusters do not exceed sample size

kmeans_constrained = KMeansConstrained(
    n_clusters=num_clusters,
    size_min=4,
    size_max=400,
    random_state=42
)

customer_df_location['Cluster'] = kmeans_constrained.fit_predict(customer_df_location)

# Plot clustering results
plt.figure(figsize=(10, 6))
plt.scatter(
    customer_df_location["CustomerEasting"],
    customer_df_location["CustomerNorthing"],
    c=customer_df_location["Cluster"],
    cmap="viridis",
    s=50,
    alpha=0.7
)
plt.scatter(
    kmeans_constrained.cluster_centers_[:, 0],
    kmeans_constrained.cluster_centers_[:, 1],
    c='red',
    marker='x',
    s=200,
    label='Centroids'
)
plt.xlabel("Customer Easting")
plt.ylabel("Customer Northing")
plt.title(f"K-Means Constrained Clustering (k={num_clusters})")
plt.legend()
plt.show()
```
