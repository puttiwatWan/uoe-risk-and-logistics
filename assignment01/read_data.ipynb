```python
candidate_df_location = pd.DataFrame({
    'CandidateNorthing': candidate_df['CandidateNorthing'],
    'CandidateEasting': candidate_df['CandidateEasting'],
    'Capacity': candidate_df['Capacity']
})
```

```python
candidate_df_location
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X = candidate_df_location[['CandidateNorthing', 'CandidateEasting', ]].copy()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Graph
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal k')
plt.show()
```

```python
optimal_k = 6  # Adjust based on the Elbow Method
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
candidate_df_location['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))

# Plot each cluster
for cluster in range(optimal_k):
    cluster_data = candidate_df_location[candidate_df_location['Cluster'] == cluster]
    plt.scatter(cluster_data['CandidateEasting'], cluster_data['CandidateNorthing'], label=f'Cluster {cluster}')

# Plot centroids (convert them back from scaled values)
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 1], centroids[:, 0], marker='X', s=200, color='black', label='Centroids')

plt.xlabel('CandidateEasting')
plt.ylabel('CandidateNorthing')
plt.title('K-Means Clustering Results')
plt.legend()
plt.show()
```

```python

# Creating the 440x440 matrix
cluster_matrix = np.zeros((440, 440), dtype=int)

# Filling the matrix: 1 if in the same cluster, 0 otherwise
for i in range(440):
    for j in range(440):
        if candidate_df.loc[i, 'Cluster'] == candidate_df.loc[j, 'Cluster']:
            cluster_matrix[i, j] = 1

# Converting to DataFrame for better visualization
cluster_df = pd.DataFrame(cluster_matrix, index=candidate_df['CandidateId'], columns=candidate_df['CandidateId'])
```
